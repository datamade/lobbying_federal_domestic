```python
def get_root_tag(zfile, fname):
    _tree = etree.parse(zfile.open(fname))
    return _tree.getroot().tag
```

```python
etree.XMLSyntaxError.print_file_and_line
```

```python
def all_files(zipfile_list, inspect_function):
    for _zfname in zipfile_list:
        _zfile = ZipFile(open(os.path.join(CLERK_CACHE, _zfname)))
        for _fname in _zfile.filelist:
            try:
                yield inspect_function(_zfile, _fname)
            except etree.XMLSyntaxError as e:
                sys.stderr.write("issue with {fn}".format(fn=_fname.filename))
                sys.stderr.write(str(e))
                continue
```

```python
tag_counts = Counter(all_files(zipfiles, get_root_tag))
```
